```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, linregress
```

```python
data = pd.read_csv('train.csv')
data = data[data.TotalBsmtSF < 5000]
data.SalePrice = np.log(data.SalePrice)
data.head(2)
```

```python
numeric_feats = list(data.dtypes[data.dtypes != 'object'].index)
numeric_feats.pop(-1)
numeric_feats.pop(0)
numeric_feats
```

```python
def plotOutliers(X, Y):
    temp = X.dropna()
    skew_val = skew(temp)
    print('Skew: ' + str(skew_val))
    fig,axs=plt.subplots(1,2,figsize=(8,3))
    plt.grid(True)
    sns.boxplot(X,orient='v',ax=axs[0])
    p = sns.regplot(X,Y,ax=axs[1])
    plt.tight_layout()
    slope, _, _, _, _ = linregress(x=p.get_lines()[0].get_xdata(),y=p.get_lines()[0].get_ydata())
    print('Slope: ' + str(slope))
```

```python
plotOutliers(data.LotFrontage, data.SalePrice)
```

```python
# Observing the regression plot after omitting the high leverage values
df = data[data.LotFrontage < 300]
plotOutliers(df.LotFrontage, df.SalePrice)
```

The data has a lot less skewness now. The effect of the the high leverage points can be seen from the difference in the slope values.

```python
# Observing the effect after logging the LotFrontage data.
df = data.copy()
df.LotFrontage = np.log(df.LotFrontage)
plotOutliers(df.LotFrontage, df.SalePrice)
```

As seen the high leverage has been dampened down, however a the skewness has reversed, with a decreased magnitude. There seem to be some leverages on the left side now. Let's see if capping them has some better effect.

```python
df = data.copy()
df.LotFrontage = np.log(df.LotFrontage)
df.loc[(df.LotFrontage < 3.25), 'LotFrontage'] = 3.25
plotOutliers(df.LotFrontage, df.SalePrice)
```

Capping does reduce the skewness by some extent. 
Inference for LotFrontage - Log the column, may or may not cap lower values with a threshold of 3.25

```python
plotOutliers(df.LotArea, df.SalePrice)
```

```python
# Observing the regression plot after omitting the high leverage values
df = data[data.LotArea < 100000]
plotOutliers(df.LotArea, df.SalePrice)
```

The skewness has decreased a lot, yet the value is substantially high. However, the influence of the high leverage points can be seen from the change in value of slope (it has doubled). Let's see the effect after logging the column.

```python
df = data.copy()
df.LotArea = np.log(df.LotArea)
plotOutliers(df.LotArea, df.SalePrice)
```

Logging seems to be more effective at reducing the leverage effect and keeps the data intact.
Inference for LotArea - Log the feature.

```python
# LotFrontage and LotArea seem to be the different aspect of the same physical feature. Let's see how the outcome
# varies with their combination
LotDetail = np.log(data.LotFrontage + data.LotArea)
plotOutliers(LotDetail, data.SalePrice)
```

The outcome shows that influential high leverage points have been neutralized. 
Inference - You can use a combination of LotFrontage and LotArea while building the model.

```python
df = data[(data['2ndFlrSF'] < 10)]
fig,axs=plt.subplots(1,2,figsize=(8,3))
plt.grid(True)
sns.distplot(df['1stFlrSF'], ax = axs[0])
sns.distplot(data['1stFlrSF'], ax = axs[1])
plt.tight_layout()
```

The cluster of data points of varying price at 0 sq ft of 2nd floor is due to the underlying dependency of SalePrice of 1stFlrSF. It can be seen that for data points with 2ndFlrSF = 0, the 1stFlrSF follows a similar distribution as the distribution followed by the entire set of 1stFlrSF. This indicates that this varying price at 0 2ndFlrSF is due to the influence by the 1stFlrSF value. How to tackle this?

```python
df = data.copy()
df['FlrSF'] = df['1stFlrSF'] + df['2ndFlrSF']
plotOutliers(df.FlrSF, df.SalePrice)
```

```python
# This new variable better represents the scenario as it eliminates the interaction problem for the 2ndFlrSF. 
# Let us now omit/dampen the high leverage points.
df = data[data.TotalBsmtSF< 5000]
df['FlrSF'] = df['1stFlrSF'] + df['2ndFlrSF']
df['FlrSF'] = np.log(df['FlrSF'])
plotOutliers(df.FlrSF, df.SalePrice)
```

```python
df = data.copy()
df['FlrSF'] = df['1stFlrSF'] + df['2ndFlrSF']
def foo():
    val = []
    for i, ar in enumerate(df['FlrSF']):
        if df.loc[i, '2ndFlrSF'] > 0:
            for j, ar1 in enumerate(df['1stFlrSF']):
                if np.abs(ar - ar1) < 25:
                    val.append(df.loc[i, 'SalePrice'] - df.loc[j, 'SalePrice'])
    return val
    
a = foo()
plt.scatter(range(len(a)), a)
```

The above plot suggests that for the same total surface area, the presence or absence of 2nd Floor doesn't cause much of a difference. Say, I get a house of 1200 sq. ft. 1st floor and 800 sq. ft. 2nd floor, and another house of 1950 sq. ft. 1st floor, then the Sale Price of the two house won't be much different.

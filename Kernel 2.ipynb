The relation seems to have an appreciable amount of non-linearity. Let's check the residual plot for
SalePrice = a(YearBlt) + b(YearBlt)^2 + c

```python
df = data.copy()
df.YearBuilt = 2016 - df.YearBuilt
df['YearBuiltSq'] = np.power(df.YearBuilt, 2)

from sklearn.linear_model import LinearRegression as LR

df2 = df.loc[:, ['YearBuilt', 'YearBuiltSq']]
model = LR().fit(df2.values, df.SalePrice.values)
preds = model.predict(df2.values)

sns.residplot(preds, df.SalePrice, lowess = True, line_kws = {'color': 'red'})
```

Problems 
- Due to recent remodelling, the year built parameter is unable to accurately correlate with SalePrice.
- Houses which have not been yet remodelled have a remodel year entry, this is faulty too.

- YearBuilt had a non-linear (quadratic) relation with SalePrice. This relation is understandable - old houses have high prices due to recent remodelling, similarly new houses cost high. However, the houses in the middle are not that new and are not yet too old to be remodelled, hence have relatively lower cost.
- For houses which have not been remodelled, we can introduce a categorical variable - isRemodelled?

```python
remodelled = data[data.YearBuilt != data.YearRemodAdd].YearRemodAdd.values

print("Number of remodelled houses: " + str(remodelled.shape[0]))
print("Fraction of houses not yet remodelled: " + str(1 - remodelled.shape[0]/data.shape[0]))

_, axis = plt.subplots(1, 2, figsize = (10,4))

axis[0].scatter(range(remodelled.shape[0]), remodelled)
sns.distplot(remodelled, ax = axis[1])
```

# Dealing with Missing Data

```python
missing_data = data.isnull().sum()/(data.shape[0])*100
missing_data = missing_data.drop(missing_data[missing_data == 0].index).sort_values(ascending = False)

print(missing_data.shape[0])

plt.figure(figsize = (15,10))
plt.xticks(rotation = 60)
sns.barplot(missing_data.index, missing_data)
```

Of these, most missing values would be due to the absence of the feature in the particular house. Let's look at them sequentially.

```python
data.PoolQC.fillna('NA', inplace = True)
data.MiscFeature.fillna('NA', inplace = True)
data.Alley.fillna('NA', inplace = True)
data.Fence.fillna('NA', inplace = True)
data.FireplaceQu.fillna('NA', inplace = True)
data.GarageType.fillna('NA', inplace = True)
data.GarageFinish.fillna('NA', inplace = True)
data.GarageQual.fillna('NA', inplace = True)
data.GarageCond.fillna('NA', inplace = True)
data.BsmtFinType2.fillna('NA', inplace = True)
data.BsmtExposure.fillna('NA', inplace = True)
data.BsmtFinType1.fillna('NA', inplace = True)
data.BsmtCond.fillna('NA', inplace = True)
data.BsmtQual.fillna('NA', inplace = True)
data.MasVnrType.fillna('NA', inplace = True)
```

```python
data.Electrical.fillna(data.Electrical.mode()[0], inplace = True)
data.MasVnrArea.fillna(0, inplace = True)

# GarageYrBlt in missing for houses which don't have any garages. We will set GarageYrBlt as -1 for these houses
data.GarageYrBlt.fillna(-1, inplace = True)
```

```python
# Assuming that houses in the same neighborhood have similar roads in front of their houses, we fill the 
# missing values of LotFrontage with mean of LotFrontages for the same neighborhood
data['LotFrontage'] = data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.mean()))
```

#### Now that we have dealt with missing values, we are going to do some feature engineering - like removing redundant ones, constructing new ones or transforming existing ones in order to improve the results of our model

# Dealing with skewed data

Usually any data with skew above 0.5 is considered as medium to highly skewed. We shall take this as the threshold for our dataset in order to transform them.

```python
skewness = data.skew()

for i in skewness.index:
    data[i] = boxcox1p(data[i], 0.25)
    
skewness2 = data.skew()
print(skewness2 - skewness)
```
